```python
FURHAT_IP = "100.69.108.47"

furhat = FurhatRemoteAPI(FURHAT_IP)
furhat.set_led(red=100, green=50, blue=50)

FACES = {'User': 'Patricia', 'Bot': 'Nazar'}
VOICES_EN = {'User': 'BellaNeural', 'Bot': 'KevinNeural'}

# Load the Keras model architecture from JSON
json_file = open(r"C:\Users\emrek\OneDrive\Masa端st端\Furhat_chatbot\Furhat-chatbot\emotiondetector.json")
model_json = json_file.read()
json_file.close()

# Load the Keras model weights
model = model_from_json(model_json)
model.load_weights(r"C:\Users\emrek\OneDrive\Masa端st端\Furhat_chatbot\Furhat-chatbot\emotiondetector.h5")

# Initialize Haar Cascade for face detection
haar_file = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(haar_file)

# Function to extract features from an image
def extract_features(image):
    feature = np.array(image)
    feature = feature.reshape(1, 48, 48, 1)
    return feature / 255.0

# Define emotion labels
labels = {0: 'angry', 1: 'disgust', 2: 'fear', 3: 'happy', 4: 'neutral', 5: 'sad', 6: 'surprise'}

```

```python
def set_persona(persona):
    furhat.set_face(character=FACES[persona], mask="Adult")
    furhat.set_voice(name=VOICES_EN[persona])
    
def set_emotion(emotion):
    if emotion == "happy":
        furhat.gesture(body=MyHappy, blocking=True)
    elif emotion == "angry":
        furhat.gesture(body=MyAngry, blocking=True)
    elif emotion == "disgust":
        furhat.gesture(body=MyDisgust, blocking=True)
    elif emotion == "surprise":
        furhat.gesture(body=MySurprise, blocking=True)
    elif emotion == "sad":
        furhat.gesture(body=MySad, blocking=True)
    elif emotion == "fear":
        furhat.gesture(body=MyFear, blocking=True)
    elif emotion == "neutral":
        furhat.gesture(body=MyNeutral, blocking=True)
```

```python
def bsay(line, emotion=None):
    if emotion:
        set_emotion(emotion)
    furhat.say(text=line, blocking=True)
```

```python
webcam = cv2.VideoCapture(0)
# Main loop for video capture and emotion prediction
while True:
    ret, frame = webcam.read()
    if not ret:
        print("Error reading frame from the webcam.")
        break

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)

    try:
        for (x, y, w, h) in faces:
            # Extract the face region
            face_image = gray[y:y + h, x:x + w]
            
            # Draw a rectangle around the face
            cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2) 

            # Resize the face image to 48x48 pixels
            face_image = cv2.resize(face_image, (48, 48))

            # Extract features and preprocess for emotion prediction
            img = extract_features(face_image)

            # Use the pre-trained model to predict the emotion
            pred = model.predict(img)

            # Get the predicted emotion label
            prediction_label = labels[pred.argmax()]
            
             # Display the predicted emotion label on the frame
            cv2.putText(frame, prediction_label, (x - 10, y - 10), cv2.FONT_HERSHEY_COMPLEX_SMALL, 2, (0, 0, 255))
            
            #respond_to_emotion(prediction_label.lower())
            # Respond to the detected emotion
            continue_processing = respond_to_emotion(prediction_label.lower())

            # Break out of the loop if the flag is set to False
            if continue_processing == False:
                break
            

            # Set Furhat's persona and gesture based on detected emotion
            set_persona('Bot')  # Assuming Bot persona for Furhat
            set_emotion(prediction_label.lower())

        # Display the output frame
        cv2.imshow("Facial Emotion Recognition", frame)

        # Break the loop if 'Esc' key is pressed
        key = cv2.waitKey(1) & 0xFF
        if key == 27:
            break

    except cv2.error:
        pass
    
# Release the webcam and close all windows
webcam.release()
cv2.destroyAllWindows()

# if __name__ == '__main__':
#     # Start the conversation with a welcome message
#     bsay("Hello! I'm Furhat, your friendly bartender. How are you feeling today?")

#     # Simulate a conversation loop
#     while True:
#         user_input = input("You: ")
        
#         # Analyze user input or perform natural language processing if needed
        
#         # Example rules for Furhat's responses based on user input
#         if "happy" in user_input.lower():
#             bsay("It's wonderful to hear that you're happy!")
#         elif "angry" in user_input.lower():
#             bsay("I'm sorry to hear that you're feeling angry. Is there anything specific bothering you?")
#         elif "tell a joke" in user_input.lower():
#             bsay("Sure, here's a joke: Why don't scientists trust atoms? Because they make up everything!")
#         elif "bye" in user_input.lower():
#             bsay("Goodbye! If you need anything, feel free to come back.")
#             break
#         else:
#             bsay("I'm here to chat! Feel free to share your thoughts or ask me anything.")

        # You can continue adding more rules based on the conversation context and user input

# if __name__ == '__main__':
   
#     bsay("I'm responding with happiness emotion", emotion="happy")
#     sleep(1)
#     bsay("I'm responding with angry emotion", emotion="angry")
#     sleep(1)
#     bsay("I'm responding with disgust emotion", emotion="disgust")
#     sleep(1)
#     bsay("I'm responding with surprise emotion", emotion="surprise")
#     sleep(1)
#     bsay("I'm responding with sad emotion", emotion="sad")
#     sleep(1)
#     bsay("I'm responding with fear emotion", emotion="fear")
#     sleep(1)
#     bsay("I'm responding with neutral emotion", emotion="neutral")
```
